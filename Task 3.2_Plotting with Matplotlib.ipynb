## <a id="labbookchecklist"></a> X-ray diffraction experiment 

This notebook conatins the data for a X-ray diffraction experiment. It shows the countrate against twice the diffraction angle for the two materials, Lithium flouride and sodium chloride. 

```python
#This line just makes the graph interactive 

%matplotlib notebook

#Then we import numpy and matplot so that we can use its features 

import numpy as np
import matplotlib.pyplot as plt

# Now we import our data from the two excel files 
#'unpack=True' unpacks each column into a separate array.
# Python normally treats white space as the demiliter so here we change it to a comma since that's how the data 
# in the excel file is seperated into into columns 

x1, y1 = np.loadtxt("LiF-30kV.csv", unpack=True, delimiter = ',')
x2, y2 = np.loadtxt("NaCl-30kV.csv", unpack=True, delimiter = ',')

# Now we plot both of the graphs, give it a tile, label teh axis and put a legend 

plt.figure() 
plt.plot(x1,y1,label="LiF")      
plt.plot(x2,y2,label="NaCl")  
plt.xlabel(r' 2$\theta$ (Â°) ')               
plt.ylabel(' Counts per second (s^-1) ')                 
plt.title(' EXperimental data for count rate against angle for LiF & NaCl ') 
plt.grid(True)
plt.legend(loc="best");

# This line allows us to save the graph so that we can print it etc.

plt.savefig('Count rate against angle for LiF & NaCl graph')
```

Here I used a continous line as it lets us see easily at which angles we have spikes in count rate . If we used data points, it would've been slightly harder to see which angles we have meaningful spikes. Same thing with a dotted or dashed line. 
I didn't choose any colours for the lines so python chose them. However these colours are fine since they are easily distinguishable from each other and make both the plots readable even when on the same graph.
