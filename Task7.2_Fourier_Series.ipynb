The code cell below contains the start of a function that needs to calculate the Fourier series for a square wave, i.e. it codes the mathematical function
$$f_\text{square}(x) = \frac{4}{\pi} \sum_{n = 1, 3, 5, \ldots}^{\infty} \frac{1}{n} \sin\left(\frac{n \pi x}{L}\right) \qquad \qquad (1)$$ 
for all $n$ up to and including $n = $ <tt>n_max</tt> (as we will need to truncate our series at some finite point rather than continuing all the way to $n = \infty$). The docstring has already been written for you.

```python
def fourier_sum(x,n_max):
    '''Creates the Fourier series for square wave.
    Inputs:
    x:       array of x-values or single x-value
    n_max:   maximum value of n in summation from eqn(1) above
    Depends on:
    L:       global variable determining period of square wave
    Returns:
    fourier: array of y-values or single y-value for Fourier square wave
    
    '''
    #Define parameteres
    answer = 0
    n = 1
    
    #Loop i upto and including n_max, adjusting the sequence each time 
    while n <= n_max :
        answer = answer + (1/n * np.sin(n * np.pi * x / L))
        n = n + 2 
    
    #Return answer
    return (4/np.pi * answer)
```

```python
#Define all the fourier functions we want to find 
f1 = fourier_sum(x,1)
f2 = fourier_sum(x,2)
f3 = fourier_sum(x,3)
f4 = fourier_sum(x,4)
f5 = fourier_sum(x,5)
f10 = fourier_sum(x,10)
f20 = fourier_sum(x,20)
f50 = fourier_sum(x,50)
f100 = fourier_sum(x,100)
f200 = fourier_sum(x,200)
f500 = fourier_sum(x,500)
```

```python
#Plot the graphs we are interested in 

plt.figure() 
plt.plot(x,f1,label="n=1")      
plt.plot(x,f5,label="n=5")
plt.plot(x,f20,label="n=20")
plt.plot(x,f100,label="n=100")

plt.xlabel('x')               
plt.ylabel('y')                 
plt.title(' Fourier series for increasing values of n ') 
plt.grid(True)
plt.legend(loc="best"); 
```

As n_max increases, the function becomes more like square wave, with small oscilations at the peaks and troughs of the graphs that slowly become smaller and unnoticable.


```python
plt.figure() 
plt.plot(x,f1,label="n=1")      
plt.plot(x,f3,label="n=3")      

plt.xlabel('x')               
plt.ylabel('y')                 
plt.title(' Fourier series for odd values of n ') 
plt.grid(True)
plt.legend(loc="best"); 

plt.figure()      
plt.plot(x,f2,label="n=2")      
plt.plot(x,f4,label="n=4")


plt.xlabel('x')               
plt.ylabel('y')                 
plt.title(' Fourier series for even values of n ') 
plt.grid(True)
plt.legend(loc="best"); 
```

Suprisingly, odd and even values of n in which the odd and even pair are consecutive have the exact same graph. However odd and even numbers of of n_max have different graphs.

```python
plt.figure() 
plt.plot(x,f100,label="n=100")
plt.plot(x,f200,label="n=200")
plt.plot(x,f500,label="n=500")

plt.xlabel('x')               
plt.ylabel('y')                 
plt.title(' Fourier series for large values of n ') 
plt.grid(True)
plt.legend(loc="best"); 
```

For very big numbers of n_max, past 100, the the graph is very square like. If we used even bigger numbers, we would see pretty much a square wave.

```python
plt.figure() 
plt.plot(x,f1,label="n=1")      
plt.plot(x,f5,label="n=5")
plt.plot(x,f20,label="n=20")
plt.plot(x,f50,label="n=50")
plt.plot(x,f100,label="n=100")


plt.xlabel('x')               
plt.ylabel('y')                 
plt.title(' Fourier series for different values of n ') 
plt.grid(True)
plt.legend(loc="best"); 
```

As n_max increases, the function changes more rapidly at the peaks and troughs of the graph. The gradient changes very quickly and and the smaller oscilattions become more frequent and smaller in amplitude, tending towards being flat. This is due to the 1/n part of the equation, 'damping' the oscilations.
